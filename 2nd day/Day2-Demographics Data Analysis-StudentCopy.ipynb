```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
# Ignore all the warnings
warnings.filterwarnings('ignore')
```

```python
"""
We have been asked to prepare a ML solution to predict BirthRate
in a country by accepting InternetAccess as input
"""
```

```python
DF = pd.read_csv(r"DemographicData.csv")
```

```python
DF.columns = (['CountryName', 'CountryCode', 'BirthRate', 'InternetAccess',
       'IncomeGroup'])
DF.head()
```

### ploting distribution of BirthRate using boxplot()

```python
sns.boxplot(data=DF,x="IncomeGroup",y="BirthRate")
# We are able to use matplotlib commands with the seaborn library
plt.xticks(rotation=55)
plt.title("BirthRate Distribution")
plt.show()
```

```python
# Describing the birthrate by only using the "High income" group
DF[DF.IncomeGroup=="High income"].describe()[["BirthRate"]]
```

```python
DF[DF.IncomeGroup=="High income"].sort_values("BirthRate")[["BirthRate"]]
```

# Mean & Median

```python
print("---------High Income Group---------")
#print(f"Mean {DF[DF.IncomeGroup=="High income"]["BirthRate"].mean()}\nMedian {DF[DF.IncomeGroup=="High income"]["BirthRate"].median()}")
print("Mean ",DF[DF.IncomeGroup=="High income"]["BirthRate"].mean())
print("Median ",DF[DF.IncomeGroup=="High income"]["BirthRate"].median())
print("---------Lower Income Group---------")
#print(f"Mean {DF[DF.IncomeGroup=="Low income"]["BirthRate"].mean()}\nMedian {DF[DF.IncomeGroup=="Low income"]["BirthRate"].median()}")
print("Mean ",DF[DF.IncomeGroup=="Lower middle income"]["BirthRate"].mean())
print("Median ",DF[DF.IncomeGroup=="Lower middle income"]["BirthRate"].median())
```

### ploting distribution of BirthRate using swarmplot()

```python
sns.swarmplot(data=DF,x="IncomeGroup",y="BirthRate")
plt.xticks(rotation=55)
plt.show()
```

### boxplot() & swarmplot()

```python
# The values in the figure size are in inches (height,width), its adjusted to the screen size
plt.rcParams["figure.figsize"]=(10,8)
sns.boxplot(data=DF,x="IncomeGroup",y="BirthRate",color="lightgrey")
sns.swarmplot(data=DF,x="IncomeGroup",y="BirthRate")
plt.show()
```
